```python
# prompt: python code for complete Linear Regression  using the csv file

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv('/content/sample_data/imputed_datasetknn.csv')

# Extract the input and output variables
X = df.iloc[:, 0].values.reshape(-1, 1)
y = df.iloc[:, 1].values

# Create a linear regression model
model = linear_model.LinearRegression()

# Train the model
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='red')
plt.plot(X, y_pred, color='blue')
plt.show()

cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize=(8,6))
m1 = plt.scatter(X, y, color=cmap(0.9), s=10)
m2 = plt.scatter(X, y, color=cmap(0.5), s=10)
plt.plot(X, y_pred, color='black', linewidth=2, label="Prediction")
plt.show()
```
