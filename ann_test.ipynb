```python
# prompt: python code for data prediction in csv file using the ANN

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error

# Load the dataset
dataset = pd.read_csv('/content/sample_data/source12.csv')

# Handle missing values (replace with mean for simplicity - consider more sophisticated methods)
dataset.fillna(dataset.mean(), inplace=True)  # Fill NaNs with column means

# Split the dataset into input (X) and output (Y) variables
# The -1 index in iloc selects only the last column for Y
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values

# Normalize the data
sc = MinMaxScaler(feature_range=(0, 1))
X = sc.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)

# Create the ANN model
# Adjust the input_shape to match the actual number of features in X
model = Sequential()
model.add(Dense(units=6, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, Y_train, epochs=100, batch_size=32)

# Make predictions
Y_pred = model.predict(X_test)

# Print the predictions
print(Y_pred)

# Step 7: Evaluate the Model
mse = mean_squared_error(Y_test, Y_pred)
print(f"Mean Squared Error: {mse}")

# Optional: Save the model
model.save('/content/sample_data/ann_model.keras')
print("Model saved as 'ann_model.keras'")
```
